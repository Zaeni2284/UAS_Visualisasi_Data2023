```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import PercentFormatter
```

 menapilkan kebutuhan bulanan

```python
df = pd.read_csv('/content/kebutuhan_per_bulan_2.csv')
df
```

SCATER PLOT 1 KEBUTUHAN PERTAHUN BERDASARKAN SEMUA PROVINSI

```python

plt.figure(figsize=(12, 8))
plt.scatter(df['tahun'], df['Propinsi'], s=50, c='magenta', alpha=0.7)

for i in range(len(df)):
    plt.annotate(df['kebutuhan'][i], (df['tahun'][i], df['Propinsi'][i]), ha='center', va='bottom', fontsize=8)

plt.title('Plot 1 - Kebutuhan Per Tahun Berdasarkan Seluruh Provinsi', fontsize=18)
plt.xlabel('tahun', fontsize=12)
plt.ylabel('Propinsi', fontsize=12)

plt.xticks(rotation=0, fontsize=8)

plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

SCATER PLOT 2 KEBUTUHAN PERTAHUN BERDASARKAN PROVINSI JAWA BARAT

```python

data_filtered_jabar = df[df['Propinsi'] == 'JAWA BARAT']

plt.scatter(data_filtered_jabar['tahun'], data_filtered_jabar['kebutuhan'], c=data_filtered_jabar.index, cmap='rainbow', s=50, alpha=0.7)

plt.title('Scatter Plot - Kebutuhan per Tahun (Jawa Barat)')
plt.xlabel('Tahun')
plt.ylabel('Kebutuhan')

cbar = plt.colorbar()
cbar.set_label('Data Index')

plt.show()
```

Sacatter 3 - Menampilkan data transaksi

```python
df_transaksi = pd.read_csv("/content/transaksi2.csv", names=['date', 'id_product', 'quantity', 'id_penjualan'])
df_transaksi
```

```python
df_transaksi['date'] = pd.to_datetime(df_transaksi['date'])
df_transaksi['year'] = df_transaksi['date'].dt.year
df_transaksi['month'] = df_transaksi['date'].dt.month_name()
df_clean = df_transaksi.loc[df_transaksi['year'] == 2017]
df_clean

plt.title('PLOT 3 - MENAMPILKAN DATA TRANSAKSI')
plt.scatter(df_clean['date'], df_clean['quantity'])

plt.xlabel('Date')
plt.ylabel('Quantity')

plt.show()
```

Sacatter 4 - Transaksi pada tahun 2018

```python
df_transaksi['date'] = pd.to_datetime(df_transaksi['date'])
df_transaksi['year'] = df_transaksi['date'].dt.year
df_transaksi['month'] = df_transaksi['date'].dt.month_name()
df_clean = df_transaksi.loc[df_transaksi['year'] == 2017]
# Melakukan pengelompokan dan penghitungan
plt.title('PLOT 4 - MENAMPILKAN DATA TRANSAKSI TAHUN 2017')
df_count = df_clean.groupby('month').count()
# Menampilkan scatter plot
plt.scatter(df_count['date'], df_count.index)
```
